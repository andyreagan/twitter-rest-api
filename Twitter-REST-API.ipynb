```python
tweetIDs = [x for x in range(510,520)]
import json
import ujson
import codecs
f = codecs.open('/Users/andyreagan/play/510-519.ujson','w','utf8')
for tweetID in tweetIDs:
    print 'getting tweet ID ' + str(tweetID)
    try:
        tweet = twitter.show_status(id=tweetID)
        #json.dump(tweet,f,ensure_ascii=False)
        ujson.dump(tweet,f,ensure_ascii=False)
        f.write('\n')
        print 'success'
    except TwythonError as e:
        print e
f.close()
```
